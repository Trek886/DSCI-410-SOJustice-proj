```python
import pandas as pd
```

```python
call_files = ['class_data_2021.csv', 'class_data_2022.csv', 'class_data_2023.csv', 'class_data_2024.csv', 'class_data_2025.csv']
calls = pd.concat([pd.read_csv(f) for f in call_files], ignore_index=True)
```

```python
calls = calls[calls['primeunit'] == 'CAHOOT']
calls.head()
```

```python
calls_time = calls[['calltime']]
calls.head()
```

```python
calls['date'] = pd.to_datetime(calls_time['calltime']).dt.date
calls.head()
```

```python
call_vol = calls.groupby('date').size().reset_index(name='call_volume')
call_vol.head()
```

```python
weather_files = ['Eugene, OR, United States... 2021-01-01 to 2023-08-09.csv', 'Eugene Oregon 2023-08-10 to 2025-04-25.csv']
weather = pd.concat([pd.read_csv(f) for f in weather_files], ignore_index=True)
```

```python
weather['date'] = pd.to_datetime(weather['datetime']).dt.date
weather.head()
```

```python
merged = pd.merge(call_vol, weather, on='date', how='inner')
merged.head()
```

```python
merged.info()
```

```python
cols_to_drop = [
    'source', 'date', 'name', 'dew', 'humidity', 'precipprob', 'precipcover',
    'windspeedmean', 'windspeedmin', 'winddir', 'sealevelpressure',
    'solarradiation', 'solarenergy', 'uvindex'
]
cleaned = merged.drop(columns=cols_to_drop)
cleaned.head()
```

```python
cleaned.info()
```

```python
extremes = (
    (cleaned['tempmax'] > 97) |
    (cleaned['tempmin'] < 25) |
    (cleaned['windspeed'] > 25) |
    (cleaned['precip'] > 1) |
    (cleaned['snow'] > 1)
)
extreme_days = cleaned[extremes]
non_extreme_days = cleaned[~extremes]
```

```python
extreme_days.head()
```

```python
non_extreme_days.head()
```

```python
extreme_days.to_csv('extreme_days.csv', index=False)
non_extreme_days.to_csv('non_extreme_days.csv', index=False)
cleaned.to_csv('merged.csv', index=False)
```
